We start by deriving an expression for the bond prices $P(t,T)$. In the slides on HJM framework, it is stated that the bond prices are explicit in the forward rates by the following relationship:

$$
P(t,T) = \exp\{-\int_{t}^{T}f(t,u)ds\}
$$

The explicit bond prices are given as:

\begin{align}
    P(0,T) &= \exp\{-\int_{0}^{T} f(0,u) du\} = \exp\{-T \cdot 0.04\} \\
    P(\omega_{1}, 1, T) &= \exp\{-\int_{1}^{T} f(\omega_{1}, 1,u) du\} = \exp\{-T \cdot 0.06 + 0.06\} = \exp\{-0.06 \cdot (T - 1)\} \\
    P(\omega_{2}, 1, T) &= \exp\{-\int_{1}^{T} f(\omega_{2}, 1,u) du\} = \exp\{-T \cdot 0.02 + 0.02\} = \exp\{-0.02 \cdot (T - 1)\}
\end{align}

We can establish a system of linear equations based on the above.

$$
A = \begin{bmatrix} 
        P(0,1) & P(0,2) & P(0,3) \\ 
        P(\omega_{1}, 1, 1) & P(\omega_{1}, 1, 2) & P(\omega_{1}, 1, 3) \\
        P(\omega_{2}, 1, 1) & P(\omega_{2}, 1, 2) & P(\omega_{2}, 1, 3) \\
     \end{bmatrix} \quad
b = \begin{bmatrix} 0 \\ 1 \\ 1 \end{bmatrix}
$$

Then create a function that sets up the system.

```python
import numpy as np

# Define bond prices
def P(w,t,T):
    if t == 0:
        return np.exp(-T*0.04)    
    elif w == 1 and t == 1:
        return np.exp(-0.06*(T-1))
    elif w == 2 and t == 1:
        return np.exp(-0.02*(T-1))
    else:
        return 0
    
# System of equations
A = np.array([[P(0,0,1), P(0,0,2), P(0,0,3)],
              [P(1,1,1), P(1,1,2), P(1,1,3)],
              [P(2,1,1), P(2,1,2), P(2,1,3)]])
b = np.array([[0], [1], [1]])
```

From the first $0$ it is understood that the portfolio is self-financing. The latter conditios imply a positive return in all outcomes. It can be shown that $A$ has full rank by showing that the determinant is non-zero.

```python
np.linalg.det(A)
```

Equivalently, just show that it has full rank.

```python
np.linalg.matrix_rank(A) == A.shape[0]
```

Hence, there exists an unique solution for the weights q such that a risk-free profit is yielded at time $t = 1$.

```python
q = np.linalg.solve(A,b)
```

Start by defining functions to get the bond prices, when the short rate follows a Vasicek-type process.

```python
from scipy.stats import norm

def affineBondPrices(TtM, rt=0.05, A=lambda TtM : 1, B=lambda TtM : 1):
    return np.exp(-A(TtM) - B(TtM)*rt)

def P_Vasicek(t,T, rt=0.05, kappa=0.01, sigma=0.02, theta=0.04):
    B=lambda TtM : (1 - np.exp(-kappa*(TtM)))/kappa
    A=lambda TtM : -((theta - (sigma**2)/(2*kappa**2))*(B(TtM) - TtM) - ((sigma**2)/(4*kappa))*(B(TtM)**2))
    return affineBondPrices(T-t, rt=rt, A=A, B=B)
```

```python
# Parameters and initial condition
r0 = 0.06
kappa = 0.86
theta = 0.08
sigma = 0.01

# Maturities
T = [1/4,1/2]

# Bond prices
P1 = P_Vasicek(0,T[1],rt=r0,kappa=kappa,theta=theta,sigma=sigma)
P0 = P_Vasicek(0,T[0],rt=r0,kappa=kappa,theta=theta,sigma=sigma)

# Strike price
K = P1/P0
```

```python
def BondEuropeanCallOption(T1,T0,K,*params):
    
    # Parameters and initial condition
    r0 = params[0]
    kappa = params[1]
    theta = params[2]
    sigma = params[3]
    
    # Calculate d1 and d2
    volIntegral = ((sigma**2)/(kappa**2))*((np.exp(-kappa*T0) - np.exp(-kappa*T1))**2)*(np.exp(2*kappa*T0) - 1)/(2*kappa)
    P1 = P_Vasicek(0,T[1],rt=r0,kappa=kappa,theta=theta,sigma=sigma)
    P0 = P_Vasicek(0,T[0],rt=r0,kappa=kappa,theta=theta,sigma=sigma)
    d1 = (np.log(P1/(K*P0)) + 0.5*volIntegral)/np.sqrt(volIntegral)
    d2 = (np.log(P1/(K*P0)) - 0.5*volIntegral)/np.sqrt(volIntegral)
                                                       
    # Return price
    p_call = P1*norm.cdf(d1) - K*P0*norm.cdf(d2)
    return p_call
    
```

For an at-the-money (ATM), the call and put options are equal in value.

```python
p_call_analytical = BondEuropeanCallOption(T[1],T[0],K,r0,kappa,theta,sigma)
print(r"Price of call options on T1-bond with expiration on T0: {} basis points".format(round(p_call_analytical*10000,2)))
```

Finally, as it is usually hard to obtain a closed-form solution, we can try to estimate the result with a Monte Carlo approach.

In general, the time $t$ price of a $T$-claim $X$ is



under the risk-neutral measure $\mathbb{Q}$. We start by using Euler-Maruyama approximation to the short rate process.

```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
import time
from statsmodels.graphics.tsaplots import plot_acf
from scipy import linalg
import random

def EulerMaruyama(f, g, t, dB, x0 = 0, transpose = True):
    
    if transpose:
        dB = np.transpose(dB)
    
    # Number of simulations
    n_sim = dB.shape[0]
    
    # Setup returned solution
    X = np.zeros([n_sim, len(t)])
    X[:,0] = x0
    dt = np.diff(t)
    
    for i in range(0, len(t) - 1):
        
        X[:,i+1] = X[:,i] + f(X[:,i]) * dt[i] + g(X[:,i]) * dB[:, i]
        
    return(X)

def BrownianMotion(t = 100, t_0 = 0, delta_t = 0.001, B_0 = 0, n_sim = 10, plot = False, PRINT = False):
    
    start = time.time()
    
    size = math.ceil((t - t_0)/delta_t)
    time_points = [t_0] + [delta_t for i in range(0, size)]
    
    column_names = ['Time'] + ['Simulation {}'.format(i) for i in range(1, n_sim+1)];
    time_array = np.cumsum(time_points);
    
    # Simulation
    simulation_array = np.cumsum(np.sqrt(delta_t) * norm.ppf(np.random.rand(size + 1,n_sim)), axis = 0);
    simulation_array[0,:]  = B_0;
    
    # Then we can make a Pandas data frame
    df = pd.DataFrame(np.column_stack([time_array, simulation_array]));
    df.columns = column_names;
    
    end = time.time()
    if PRINT:
        print("\nTime to run simulations: {}s \n".format(end - start))
        print("The output has {} rows and {} columns.".format(df.shape[0], df.shape[1]))
        print("The total number of elements is {}.".format(df.shape[0]*df.shape[1]))
    
    if plot:
        plt.figure(figsize=(12,6), dpi = 1000)
        plt.xlabel('t', fontsize = 12)
        plt.ylabel('$B_{t}$', fontsize = 12) 
        plt.title('Simulation of Brownian Motion')
        plt.plot(df.iloc[:,0].values, df.iloc[:,1:].values);
    
    return(df)
```

Based on the short-rate, we can obtain the time $t$ $T$-bond prices for $ 0 \leq t \leq T$.

```python
def P(s,S,t=t,r=rates):
    
    # Get indices
    start = np.argmin(np.abs(t-s))
    end = np.argmin(np.abs(t-S))

    # Approximate integral
    integral = 0
    for i in range(len(t[start:end])):
        integral += r[:,start+i]*(t[start+i+1]-t[start+i])

    # Save result
    bondPrices = np.exp(-integral)
        
    # Return result
    return bondPrices
```
