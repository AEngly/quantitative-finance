### 2)

We start by construction the C matric and the p vector.

```python
n = maturities.shape[0]
m = df_merged.shape[0]
p = np.zeros((n,1))
```

These prices are notional.

Insert the price vector for manual inspection.

```python
# Insert zeros
df_merged['Prices'] = np.zeros((39,1))

# Looping
index = 0
for i in range(df_merged.shape[0]):
    if not pd.isna(df_merged.loc[i,'Quote']):
        df_merged.loc[i,'Prices'] = p[index]
        index += 1
```

Then construct the sparse cash flow matrix. The cash flows correspond to all the dates we wish to determine the discount curve for.

```python
C = np.zeros((n,m))
```

Then we fill it out in accordance with the procedure described in the slides.

```python
# LIBOR before gap
for i in range(3):
    t0 = spotDate
    T = maturities[i]
    delta = (T - t0).days/360
    L = quotes[i]/100
    C[i,i] = 1 + delta*L
    
# LIBOR after gap
t0 = spotDate
T = maturities[3]
delta = (T - t0).days/360
L = quotes[3]/100
C[3,4] = 1 + delta*L

# Future 1
T1 = tenor[3]
T2 = maturities[4]
delta = (T2 - T1).days/360
F = (100 - quotes[4])/100
C[4,3] = -1
C[4,5] = 1 + delta*F

# Future 2
T1 = maturities[4]
T2 = maturities[5]
delta = (T2 - T1).days/360
F = (100 - quotes[5])/100
C[5,5] = -1
C[5,6] = 1 + delta*F

# Future 3
T1 = maturities[5]
T2 = maturities[6]
delta = (T2 - T1).days/360
F = (100 - quotes[6])/100
C[6,6] = -1
C[6,8] = 1 + delta*F

# Future 4
T1 = maturities[6]
T2 = maturities[7]
delta = (T2 - T1).days/360
F = (100 - quotes[7])/100
C[7,8] = -1
C[7,9] = 1 + delta*F

# Swap 1
K = quotes[8]/100
T1 = spotDate
T2 = tenor[7]
delta = (T2 - T1).days/360
C[8,7] = delta*K
T1 = tenor[7]
T2 = tenor[10]
delta = (T2 - T1).days/360
C[8,10] = 1 + delta*K

# Swap 2
K = quotes[9]/100
T1 = spotDate
T2 = tenor[7]
delta = (T2 - T1).days/360
C[9,7] = delta*K
T1 = tenor[7]
T2 = tenor[10]
delta = (T2 - T1).days/360
C[9,10] = delta*K
T1 = tenor[10]
T2 = tenor[11]
delta = (T2 - T1).days/360
C[9,11] = 1 + delta*K

# Remaining swaps
swapDates = np.array([tenor[7]] + [tenor[i] for i in range(10, len(tenor))])
deltas = np.array([(swapDates[0] - spotDate).days/360] + [(swapDates[i]-swapDates[i-1]).days/360 for i in range(1,len(swapDates))])
indices = np.array([7,10,11] + [i for i in range(12,len(tenor))])
remainingSwapsMaturity = maturities[10:]

for index, maturityDate in enumerate(remainingSwapsMaturity):
    lastIndex = np.argwhere(tenor == maturityDate)[0,0]
    lastDeltaIndex = np.argwhere(indices == lastIndex)[0,0]
    K = quotes[10+index]/100
    for i in range(lastDeltaIndex):
        C[10+index, indices[i]] = deltas[i]*K
    C[10+index, lastIndex] = 1 + deltas[lastDeltaIndex]*K
        
# Make a loop to fill out the remaining
```

Then we can continue with the psudo-inverse method.

```python
M = np.eye(C.shape[1])
M[1:,:-1] = M[1:,:-1] - np.eye(C.shape[1]-1)
```

Then build W.

```python
W = np.diag([1/np.sqrt((tenor[0]-spotDate).days/360)] + [1/np.sqrt((tenor[i+1]-tenor[i]).days/360) for i in range(len(tenor)-1)])
aux = np.zeros((C.shape[1],1))
aux[0] = 1
```

```python
RHS = p - np.matmul(np.matmul(C,np.linalg.inv(M)),aux)
LHS = np.matmul(C, np.matmul(np.linalg.inv(M),np.linalg.inv(W)))
```

Then define A.

```python
A = np.matmul(np.matmul(C, np.linalg.inv(M)), np.linalg.inv(W))
```

Then find delta.

```python
delta = np.matmul(np.matmul(np.transpose(A), np.linalg.inv(np.matmul(A, np.transpose(A)))), p - np.matmul(np.matmul(C, np.linalg.inv(M)), aux)) 
```

Then recover the discount curve.

```python
d = np.matmul(np.linalg.inv(M), np.matmul(np.linalg.inv(W), delta) + aux)
```

Compute answer.

```python
ZCB_2 = d[-1,0]
ZCB_1 = d[-2,0]
delta = (tenor[-1] - tenor[-2]).days/360
print("F(0,2041,2042) = {}%".format(round((ZCB_1/ZCB_2 - 1)*(1/delta)*100, 2)))
```

### 3)

Insert data from quiz.

```python
T = np.array([2,3,4,5,7,10,20,30])
y = np.array([-0.79, -0.73, -0.65, -0.55, -0.33, -0.04, 0.54, 0.73])
N = len(T)
ZCB = np.array([np.exp(-T[i]*y[i]/100) for i in range(N)])
alpha = 0.1
```

Define key functions.

```python
def h_derivative(u,i):
    return max(T[i] - u, 0)

def h(u,i):
    return T[i]*(1 + min(T[i],u)) - (1/2)*(min(T[i], u)**2)

def h_integral(S,i):
    return T[i-1]*min(S,T[i-1])*(1 + (1/2)*min(S,T[i-1])) - (1/6)*(min(T[i-1],S)**3) + max(S-T[i-1],0)*(T[i-1]*(1 + T[i-1]) - (1/2)*(T[i-1]**2))

def innerH(i,j):
    return -(min(T[i],T[j])**3)/6 + T[i]*T[j]*(1 + min(T[i], T[j])/2)

def aux(i,j):
    if i == j:
        return 1
    else:
        return 0
```

Solve the system of equations.

```python
# RHS
Z = alpha*np.array([y[i]*T[i] for i in range(len(T))]).reshape(N,1)
RHS = np.vstack([np.array([[0]]),Z])

# LHS
H = np.array([[alpha*innerH(i,j) + aux(i,j) for j in range(N)] for i in range(N)])
A = np.array([0] + [alpha*T[i] for i in range(N)]).reshape(N+1,1)
LHS = np.hstack([A, np.vstack([T,H])])

# Solve linear system
beta = np.linalg.solve(LHS,RHS)

# Define forward curve with basis functions
def f(u):
    return beta[0] + sum([beta[i+1]*h(u,i) for i in range(N)])

# Define yield curve
def yieldCurve(S):
    return beta[0] + (1/S)*sum([beta[i]*h_integral(S,i) for i in range(1,N+1)])
```

Then we visualize the results.

```python
x = np.linspace(0.01,30,1000)
z = np.array([yieldCurve(T) for T in x])
forwardCurve = np.array([f(T) for T in x])

plt.figure(figsize=(12,12))
plt.plot(x,z, label="Cubic Spline")
plt.plot(x,forwardCurve, label="Instantaneous Forward Curve")
plt.scatter(T,y, label="Market Quotes")
plt.legend(fontsize=14)
plt.grid()
plt.xlabel("Maturity", fontsize=14)
plt.ylabel("Percentage [%]", fontsize=14)
plt.show()
```

### 4)

Load data.

```python
df = pd.read_excel('Data/SwissGovYields.xls')
X = df.to_numpy()[2:,1:]
X = np.diff(X, axis=0)
```

Estimate the covariance matrix.

```python
Sigma = np.cov(X.astype(np.float64), rowvar=False)
```

Then find the eigenvalues and the eigenvectors.

```python
lambdas, A = np.linalg.eig(Sigma)
```

It is informative to look at the factor loadings. We start by recentering and rotating the data with a principal components transform.

```python
mu = np.mean(X, axis=0).astype(np.float64)
Y = np.matmul(np.transpose(A), np.transpose(X - mu))
```

Check that these are uncorrelated.

```python
np.round(np.sum(np.cov(Y.astype(np.float64), rowvar=True), axis=1), 2)
```

Compare to eigenvalues.

```python
np.round(lambdas, 2)
```

They are definitely uncorrelated, and each eigenvalue represents the variance of the empirical principal components. Lastly, let's visualize the first three loadings (i.e., eigenvectors).

```python
x = np.array([2,3,4,5,7,10,20,30])
a1 = A[:,0]
a2 = A[:,1]
a3 = A[:,2]

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Define a color palette
colors = ['#000000', '#8B0000', '#FF4500']

plt.figure(figsize=(12, 8))

# Plot each line with the chosen color palette and line style
plt.plot(x, a1, label=r"$a_{1}$ (level)", color=colors[0], linewidth=2, marker='o')
plt.plot(x, a2, label=r"$a_{2}$ (slope)", color=colors[1], linewidth=2, marker='s')
plt.plot(x, a3, label=r"$a_{3}$ (curvature)", color=colors[2], linewidth=2, marker='D')

# Customize the legend
plt.legend(fontsize=14, frameon=False)

# Customize grid lines and style
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Add labels with appropriate font size
plt.xlabel("Maturity", fontsize=14, labelpad=10)
plt.ylabel("Percentage [%]", fontsize=14, labelpad=10)

# Set ticks parameters
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Customize the spines (borders) of the plot
for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

# Enhance the remaining spines
plt.gca().spines['left'].set_linewidth(1.2)
plt.gca().spines['bottom'].set_linewidth(1.2)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
print("Variance explained by first two principal components = {}%".format(round(100*sum(lambdas[:2])/sum(lambdas),2)))
```
